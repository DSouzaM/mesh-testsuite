```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import sem, t
from scipy import mean
import numpy as np
import random

confidence = 0.95
```

```python
directory = '../../0-firefox/memory/'

mesh_smallest_row_count = float('inf')
malloc_smallest_row_count = float('inf')
jemalloc_smallest_row_count = float('inf')

number_of_iterations = 5

for i in range(number_of_iterations):
    df = pd.read_csv(directory +str(i)+'-mesh.tsv', sep='\t')
    if(df.shape[0] < mesh_smallest_row_count):
        mesh_smallest_row_count = df.shape[0]
        
for i in range(number_of_iterations):
    df = pd.read_csv(directory+str(i)+'-malloc.tsv', sep='\t')
    if(df.shape[0] < malloc_smallest_row_count):
        malloc_smallest_row_count = df.shape[0]
        
for i in range(number_of_iterations):
    df = pd.read_csv(directory+str(i)+'-jemalloc.tsv', sep='\t')
    if(df.shape[0] < jemalloc_smallest_row_count):
        jemalloc_smallest_row_count = df.shape[0]
print(mesh_smallest_row_count,malloc_smallest_row_count, jemalloc_smallest_row_count)

mesh_col_headers = []
mesh_cols = []
for i in range(mesh_smallest_row_count):
    mesh_cols.append([0] * number_of_iterations)
    
# MESH

'''
See https://stackoverflow.com/questions/6482889/get-random-sample-from-list-while-maintaining-ordering-of-items
for method of sampling
'''

for i in range(number_of_iterations):
    df = pd.read_csv(directory +str(i)+'-mesh.tsv', sep='\t')
    col = df['rss'].to_list()
    indices = random.sample(range(len(col)), mesh_smallest_row_count)
    col = [col[i] for i in sorted(indices)]
    mesh_col_headers.append("rss"+str(i))
    for j in range(mesh_smallest_row_count):
        mesh_cols[j][i] = col[j]
    
    
print(np.array(mesh_cols).shape)
df1 = pd.DataFrame(mesh_cols, columns=mesh_col_headers)

# for i in range(20):
#     df[mesh_col_headers[i]] = mesh_cols[i]
    # df1['rss'+str(i)] = df['rss']
#     print(df1.shape, df.shape)
#     print(df1['rss'+str(i)].equals(df['rss'].head(df1.shape[0])))

# for i in range(20):
#     df = pd.read_csv(directory+str(i)+'-malloc.tsv', sep='\t').sample(frac=0.7).sort_index()
#     df2['rss'+str(i)] = df['rss']

print(df1.head())
print(df1.shape)

# df2 = pd.read_csv('1-mesh.tsv', sep='\t')
# df3 = pd.read_csv('2-mesh.tsv', sep='\t')


#df1 = df1.dropna(axis=0)
df1['ci_hi'] = np.nan
df1['ci_lo'] = np.nan
df1['mean'] = df1.mean(axis=1)

print(df1.shape)


rows_as_lists = df1.drop(['ci_hi', 'ci_lo'], axis=1).values.tolist()

print(rows_as_lists[:1])
print("row list ", len(rows_as_lists))
for i in range(len(rows_as_lists)):
    data = rows_as_lists[i]
    n = len(data)
    m = np.mean(data)
    std_err = sem(data)
    h = std_err * t.ppf((1 + confidence) / 2, n - 1)
    df1.ix[i,'ci_hi'] = m+h
    df1.ix[i,'ci_lo'] = m-h
    # print(m-h, m+h)
df1['percentage'] = np.arange(df1.shape[0])
df1['percentage'] = df1['percentage'].div(df1.shape[0])
# print(df1.head())
# # print(df1.tail())



malloc_col_headers = []
malloc_cols = []
for i in range(malloc_smallest_row_count):
    malloc_cols.append([0] * number_of_iterations)
    

'''
See https://stackoverflow.com/questions/6482889/get-random-sample-from-list-while-maintaining-ordering-of-items
for method of sampling
'''

for i in range(number_of_iterations):
    df = pd.read_csv(directory +str(i)+'-malloc.tsv', sep='\t')
    col = df['rss'].to_list()
    indices = random.sample(range(len(col)), malloc_smallest_row_count)
    col = [col[i] for i in sorted(indices)]
    malloc_col_headers.append("rss"+str(i))
    for j in range(malloc_smallest_row_count):
        malloc_cols[j][i] = col[j]
    
    
print(np.array(malloc_cols).shape)
df2 = pd.DataFrame(malloc_cols, columns=malloc_col_headers)





# # MALLOC
# del df2['time']
# del df2['kernel']
# df2 = df2.dropna(axis=0)
df2['ci_hi'] = np.nan
df2['ci_lo'] = np.nan
df2['mean'] = df2.mean(axis=1)

print(df2.shape)


rows_as_lists = df2.drop(['ci_hi', 'ci_lo'], axis=1).values.tolist()

print(rows_as_lists[:3])
for i in range(len(rows_as_lists)):
    data = rows_as_lists[i]
    n = len(data)
    m = np.mean(data)
    std_err = sem(data)
    h = std_err * t.ppf((1 + confidence) / 2, n - 1)
    df2.ix[i,'ci_hi'] = m+h
    df2.ix[i,'ci_lo'] = m-h
    #print(m-h, m+h)
df2['percentage'] = np.arange(df2.shape[0])
df2['percentage'] = df2['percentage'].div(df2.shape[0])
print(df2.head())
print(df2.shape)
# # print(df1.tail())
    
    

# JEMALLOC
jemalloc_col_headers = []
jemalloc_cols = []

for i in range(jemalloc_smallest_row_count):
    jemalloc_cols.append([0] * number_of_iterations)

for i in range(number_of_iterations):
    df = pd.read_csv(directory +str(i)+'-jemalloc.tsv', sep='\t')
    col = df['rss'].to_list()
    indices = random.sample(range(len(col)), jemalloc_smallest_row_count)
    col = [col[i] for i in sorted(indices)]
    jemalloc_col_headers.append("rss"+str(i))
    for j in range(jemalloc_smallest_row_count):
        jemalloc_cols[j][i] = col[j]
    
    
print(np.array(jemalloc_cols).shape)
df3 = pd.DataFrame(jemalloc_cols, columns=jemalloc_col_headers)





# del df2['time']
# del df2['kernel']
# df2 = df2.dropna(axis=0)
df3['ci_hi'] = np.nan
df3['ci_lo'] = np.nan
df3['mean'] = df3.mean(axis=1)

print(df3.shape)


rows_as_lists = df3.drop(['ci_hi', 'ci_lo'], axis=1).values.tolist()

print(rows_as_lists[:3])
for i in range(len(rows_as_lists)):
    data = rows_as_lists[i]
    n = len(data)
    m = np.mean(data)
    std_err = sem(data)
    h = std_err * t.ppf((1 + confidence) / 2, n - 1)
    df3.ix[i,'ci_hi'] = m+h
    df3.ix[i,'ci_lo'] = m-h
    #print(m-h, m+h)
df3['percentage'] = np.arange(df3.shape[0])
df3['percentage'] = df3['percentage'].div(df3.shape[0])
print(df3.head())
print(df3.shape)
# # print(df1.tail())
    
```

```python
fig = go.Figure(layout=go.Layout(
        title=go.layout.Title(text="Comparison of memory footprint in Firefox"),
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)',
    ))

df_list = [df1,df2,df3]
allocators = ['mesh', 'malloc', 'jemalloc']
fill = ['rgb(0,100,80)', 'rgb(255,69,0)', 'rgb(128,0,128)']
fillcolour = ['rgba(0,100,80,0.2)', 'rgba(255,69,0,0.2)', 'rgba(128,0,128,0.2)']
x = df1['percentage'].to_list()
y = df1['mean'].to_list()
y_upper = df1['ci_hi'].to_list()
y_lower = df1['ci_lo'].to_list()

for i in range(len(df_list)):
    x = df_list[i]['percentage'].to_list()
    y = df_list[i]['mean'].to_list()
    y_upper = df_list[i]['ci_hi'].to_list()
    y_lower = df_list[i]['ci_lo'].to_list()
    fig.add_trace(go.Scatter(
        x=x,
        y=y,
        line=dict(color=fill[i]),
        name=allocators[i],
        mode='lines'
    ))
    
    fig.add_trace(go.Scatter(
        x=x+x[::-1], # x, then x reversed
        y=y_upper+y_lower[::-1], # upper, then lower reversed
        fill='toself',
        fillcolor=fillcolour[i],
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo="skip",
        showlegend=False
    ))

fig.show()
```
