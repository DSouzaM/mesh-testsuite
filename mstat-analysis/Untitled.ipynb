```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import sem, t
from scipy import mean
import numpy as np

confidence = 0.95
```

```python
df1 = pd.read_csv('../runlarson.sh/0-mesh.tsv', sep='\t')
df2 = pd.read_csv('../runlarson.sh/0-malloc.tsv', sep='\t')
for i in range(20):
    df = pd.read_csv('../runlarson.sh/'+str(i)+'-mesh.tsv', sep='\t')
    df1['rss'+str(i)] = df['rss']

for i in range(20):
    df = pd.read_csv('../runlarson.sh/'+str(i)+'-malloc.tsv', sep='\t')
    df2['rss'+str(i)] = df['rss']



# df2 = pd.read_csv('1-mesh.tsv', sep='\t')
# df3 = pd.read_csv('2-mesh.tsv', sep='\t')

# df1['rss1'] = df2['rss']
# df1['rss2'] = df2['rss']
#MESH
del df1['time']
del df1['kernel']
df1 = df1.dropna(axis=0)
df1['ci_hi'] = np.nan
df1['ci_lo'] = np.nan
df1['mean'] = df1.mean(axis=1)

print(df1.shape)


rows_as_lists = df1.drop(['ci_hi', 'ci_lo'], axis=1).values.tolist()

print(rows_as_lists[:3])
for i in range(len(rows_as_lists)):
    data = rows_as_lists[i]
    n = len(data)
    m = np.mean(data)
    std_err = sem(data)
    h = std_err * t.ppf((1 + confidence) / 2, n - 1)
    df1.ix[i,'ci_hi'] = m+h
    df1.ix[i,'ci_lo'] = m-h
    #print(m-h, m+h)
df1['percentage'] = np.arange(df1.shape[0])
df1['percentage'] = df1['percentage'].div(df1.shape[0])
print(df1.head())
# print(df1.tail())

# MALLOC
del df2['time']
del df2['kernel']
df2 = df2.dropna(axis=0)
df2['ci_hi'] = np.nan
df2['ci_lo'] = np.nan
df2['mean'] = df2.mean(axis=1)

print(df2.shape)


rows_as_lists = df2.drop(['ci_hi', 'ci_lo'], axis=1).values.tolist()

print(rows_as_lists[:3])
for i in range(len(rows_as_lists)):
    data = rows_as_lists[i]
    n = len(data)
    m = np.mean(data)
    std_err = sem(data)
    h = std_err * t.ppf((1 + confidence) / 2, n - 1)
    df2.ix[i,'ci_hi'] = m+h
    df2.ix[i,'ci_lo'] = m-h
    #print(m-h, m+h)
df2['percentage'] = np.arange(df2.shape[0])
df2['percentage'] = df2['percentage'].div(df2.shape[0])
print(df1.head())
# print(df1.tail())
    
    
```
